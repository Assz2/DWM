## Interpolazione
lo scopo di questo documento sarà quello di pulire il dataset con nuovi metodi sfruttiamo la funzione interpolate di pandas


```python
import os
from dotenv import load_dotenv
import pandas as pd
import numpy as np
import utils

load_dotenv()
TRAIN_SET = os.getenv("TRAIN_PATH")
TEST_SET  = os.getenv("TEST_PATH")

train = pd.read_csv(TRAIN_SET)
test = pd.read_csv(TEST_SET)

train = train.dropna(subset=['sii'])

y_train = train['sii']

y_train.shape
```

```python
import importlib
import utils as ut
importlib.reload(utils)

X_train = ut.intersect_features(train, test)
Cl_train, dropped = ut.drop_columns(X_train)
Cl_train.head()
```

```python
Cl_train_interpolate = Cl_train.interpolate(method='polynomial', order=2)   
Cl_train_interpolate.head()
```

```python
Cl_train_interpolate.info()
print("=====================================================\n shape: ", Cl_train_interpolate.shape)
```

```python
Cl_train_interpolate_converted = ut.convert_float_fill_mean(Cl_train_interpolate)
Cl_train_interpolate_converted.info()
print("=====================================================\n shape: ", Cl_train_interpolate_converted.shape)
```

```python
Cl_train_interpolate_converted.head()
```

```python
cleaned_data_export = pd.concat([Cl_train_interpolate_converted, y_train], axis=1)   

output_path = os.path.join('dataset/', 'cleaned_data_v3.csv')
cleaned_data_export.to_csv(output_path, index=False)

print(f"Cleaned dataset saved to {output_path}")
```
