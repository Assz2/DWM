```python
# I want to collapse togheter all the features that have the substring 'FGC': 
# for each row I want to sumup in the FGC feature the values of it with the value of the 'FGC-Zone' specific feature
def clash_features(train):
    FGC_features = [col for col in train.columns if 'FGC' in col] 
    FGC_features.remove('FGC-Season')
    removed_features = 0
    zone_features_to_drop = []
    #train['Physical'] = train[physical_features].sum(axis=1)

    # I have FGC-FGC_CU, FGC-FGC_CU_Zone; FGC-GSND, FGC-GSND_Zone; and it goes on. I want to sum up in the FGC-FGC_CU a (+10% of that feature)*FGC_CU_Zone for every feature that has the corrispective _Zone feature   
    for feature in FGC_features:
        zone_feature = feature + '_Zone'

        if zone_feature in train.columns:
            print(f'Feature: {feature} - Zone: {zone_feature}')
            # Update the feature with 10% of the corresponding zone feature
            train[feature] = train[feature] + (train[zone_feature] * 0.1)
            # Add the zone_feature to the drop list
            zone_features_to_drop.append(zone_feature)
            removed_features += 1

    # Drop the zone features
    train = train.drop(zone_features_to_drop, axis=1)
    return train
```

```python
train = clash_features(train)
```
